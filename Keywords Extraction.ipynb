```python
import PyPDF2
import textract

from rake_nltk import Rake
import string
from string import punctuation
from string import digits
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
import nltk
```

```python
Filename='/home/pank/Home Practice/JavaBasics-notes.pdf'

pdfFileObj=open(Filename,'rb')
pdfReader=PyPDF2.PdfFileReader(pdfFileObj)

num_pages=pdfReader.numPages
count=0
text=""


while count<num_pages:
    pageObj=pdfReader.getPage(count)
    count+=1
    text+=pageObj.extractText()
    
if text!="":
    text=text
    
else:
    text=textract.process(Filename,method='tesseract',laguage='eng')
    
print(text)    

```

```python
tokenizer = RegexpTokenizer(r'\w+')
words=tokenizer.tokenize(text)
```

```python
stop_words=stopwords.words('english')

customStopWords = set(stopwords.words('english')+list(punctuation))
w=[word for word in words if word not in customStopWords]

s=filter(lambda x: x.isalpha(), w)


freq = nltk.FreqDist(s)
print("ALL KEYWORDS OF DOCUMENT")
for key,val in freq.items():
      if (len(key))>4:        
            print (str(key) + ' : ' + str(val))
                
    
```
